# **Percentage Growth **

for my analysis kaga

```python
#na thaniya columns create panni oru analysis sollalamnu itha pana soo itha vera dataframela save panipa
import pandas as pd

# Function to calculate percentage growth
def growth(curr, prev):
    return ((curr - prev) / prev) * 100 if prev != 0 else 0

# List of crop prefixes
crop_prefixes = [
    "rice", "wheat", "kharif_sorghum", "rabi_sorghum", "sorghum",
    "pearl_millet", "maize", "finger_millet", "groundnut", "sunflower",
    "soyabean", "oilseeds", "sugarcane", "cotton"
]

# Reset the index to ensure it is consecutive integers
df = df.reset_index(drop=True)  # drop=True to avoid creating an 'index' column

# Loop through each crop and calculate overall_growth column
for crop in crop_prefixes:
    overall_growth = [None]  # First row has no previous for comparison
    for i in range(1, len(df)):
        area_growth = growth(df.loc[i, f"{crop}_area"], df.loc[i - 1, f"{crop}_area"])
        prod_growth = growth(df.loc[i, f"{crop}_production"], df.loc[i - 1, f"{crop}_production"])
        yield_growth = growth(df.loc[i, f"{crop}_yield"], df.loc[i - 1, f"{crop}_yield"])
        avg_growth = (area_growth + prod_growth + yield_growth) / 3
        overall_growth.append(round(avg_growth, 2))

    # Add the new column to the DataFrame
    df[f"{crop}_growth"] = overall_growth

# 🟢 Done! Your DataFrame now has one _overall_growth column for each crop group
```

```python
# Convert all *_growth columns to int if you want to match other int64 columns
for crop in crop_prefixes:
    growth_col = f"{crop}_growth"
    if growth_col in df.columns:
        df[growth_col] = df[growth_col].fillna(0).round().astype(int)
```

```python
#14.Impact of Area Cultivated on Production (Rice, Wheat, Maize)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import pearsonr

# Optional: Filter rows with valid values only
df_clean = df[['rice_area', 'rice_production', 'wheat_area', 'wheat_production', 'maize_area', 'maize_production']].dropna()

# 🔹 Setup for 3 plots
fig, axes = plt.subplots(1, 3, figsize=(18, 5))

# 🟩 RICE
sns.regplot(x='rice_area', y='rice_production', data=df_clean, ax=axes[0], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
axes[0].set_title('Rice: Area vs Production')
axes[0].set_xlabel('Rice Area (Ha)')
axes[0].set_ylabel('Rice Production (Tonnes)')

# 🟧 WHEAT
sns.regplot(x='wheat_area', y='wheat_production', data=df_clean, ax=axes[1], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
axes[1].set_title('Wheat: Area vs Production')
axes[1].set_xlabel('Wheat Area (Ha)')
axes[1].set_ylabel('Wheat Production (Tonnes)')

# 🟨 MAIZE
sns.regplot(x='maize_area', y='maize_production', data=df_clean, ax=axes[2], scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
axes[2].set_title('Maize: Area vs Production')
axes[2].set_xlabel('Maize Area (Ha)')
axes[2].set_ylabel('Maize Production (Tonnes)')

plt.suptitle('Impact of Area Cultivated on Production (Rice, Wheat, Maize)', fontsize=16)
plt.tight_layout()
plt.show()
```

```python
#1.Year-wise Trend of Rice Production Across States (Top 3)

mycursor.execute("""
  SELECT year, state_name, total_rice_production
FROM (
    SELECT
        year,
        state_name,
        SUM(rice_production) AS total_rice_production,
        RANK() OVER (PARTITION BY year ORDER BY SUM(rice_production) DESC) AS rank_in_year
    FROM agri
    GROUP BY year, state_name
) AS ranked
WHERE rank_in_year <= 3
ORDER BY year, rank_in_year
""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#2.Top 5 Districts by Wheat Yield Increase Over the Last 5 Years

mycursor.execute("""SELECT
    dist_code,
    dist_name,
    MAX(CASE WHEN year = (SELECT MAX(year) FROM agri) THEN wheat_yield END) AS latest_year_wheat_yield,
    MAX(CASE WHEN year = (SELECT MAX(year) FROM agri) - 5 THEN wheat_yield END) AS past_year_wheat_yield,
    (MAX(CASE WHEN year = (SELECT MAX(year) FROM agri) THEN wheat_yield END) -
     MAX(CASE WHEN year = (SELECT MAX(year) FROM agri) - 5 THEN wheat_yield END)
    ) AS wheat_yield_increase,
    (SELECT MAX(year) FROM agri) AS latest_year,
    (SELECT MAX(year) FROM agri) - 5 AS past_year
FROM agri
WHERE
    year IN ((SELECT MAX(year) FROM agri), (SELECT MAX(year) FROM agri) - 5)
GROUP BY
    dist_code, dist_name
HAVING
    COUNT(DISTINCT year) = 2  -- District should have data for both years
ORDER BY
    wheat_yield_increase DESC
LIMIT 5
""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```

```python
#3.States with the Highest Growth in Oilseed Production (5-Year Growth Rate)
#to find the last 5 year growth rate
#Growth (%) = ((Production in recent year - Production 5 years ago) / Production 5 years ago) * 100

mycursor.execute("""SELECT
    state_name,
    SUM(CASE WHEN year = (SELECT MAX(year) FROM agri) THEN oilseeds_production END) AS latest_production,
    SUM(CASE WHEN year = (SELECT MAX(year) FROM agri) - 5 THEN oilseeds_production END) AS past_production,
    ROUND(
        (SUM(CASE WHEN year = (SELECT MAX(year) FROM agri) THEN oilseeds_production END) -
         SUM(CASE WHEN year = (SELECT MAX(year) FROM agri) - 5 THEN oilseeds_production END)) * 100.0 /
         NULLIF(SUM(CASE WHEN year = (SELECT MAX(year) FROM agri) - 5 THEN oilseeds_production END), 0),
    2) AS growth_rate_percentage
FROM agri
WHERE
    year IN ((SELECT MAX(year) FROM agri), (SELECT MAX(year) FROM agri) - 5)
GROUP BY state_name
HAVING COUNT(DISTINCT year) = 2
ORDER BY growth_rate_percentage DESC
LIMIT 10
""")
out=mycursor.fetchall()
from tabulate import tabulate
print(tabulate(out,headers=[i[0] for i in mycursor.description],  tablefmt='psql'))
```
